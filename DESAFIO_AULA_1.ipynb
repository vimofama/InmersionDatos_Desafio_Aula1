1. Promedio de área de todos los inmuebles en el dataset. El top 10.

```python
import pandas as pd
#Acceder a nuestro dataset y modificar el nombre de las columnas.
inmuebles = pd.read_csv('inmuebles_bogota.csv')
columnas = {'Baños': 'Banos', 'Área': 'Area'}
inmuebles = inmuebles.rename(columns=columnas)

#Conocer los barrios barrios de nuestro dataset, estos datos los descargamos para tener la lista de los barrios.
inmuebles.Barrio.value_counts()
```

```python
import csv
import matplotlib.pyplot as plt

#Abrir el archivo con el nombre de los barrios.
archivo = open("Barrios.csv", encoding="utf8")
reader = csv.reader(archivo)
areas = []

#Se crea un arreglo de las areas de cada barrio.
for linea in reader:
    inmuebles_barrio = (inmuebles['Barrio'] == linea[0])
    barrio = inmuebles[inmuebles_barrio]
    area = barrio['Area'].mean()
    diccionario = {'Barrio': linea[0], 'Area': round(area, 2)}
    areas.append(diccionario)

#El arreglo de areas se convierte en un DataFrame que ayudara para realizar los graficos.
serie_areas = pd.DataFrame(areas)

#Ordenamos las areas de mayor a menor.
areas_ordenadas = serie_areas.sort_values('Area', ascending=False)

#Grafica los 5 primeros barrios y su area.
plt.bar(areas_ordenadas.iloc[0:5].Barrio, areas_ordenadas.iloc[0:5].Area)
```

```python
#Grafica los 5 siguientes barrios y su area.
plt.bar(areas_ordenadas.iloc[5:10].Barrio, areas_ordenadas.iloc[5:10].Area)
```

```python
#Muestra los 10 primeros barrios con su respectiva area.
areas_ordenadas.iloc[0:10]
```
